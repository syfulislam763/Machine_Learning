```python
import pandas as pd
from sklearn.datasets import load_wine
wine = load_wine()
```

```python
dir(wine)
```

```python
df = pd.DataFrame(wine.data, columns=wine.feature_names)
df.head()
```

```python
df['target'] = wine.target
df.sample(5)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(['target'],axis='columns'), df.target, test_size=0.25)
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB
g_model = GaussianNB()
m_model = MultinomialNB()
```

```python
g_model.fit(X_train, y_train)
```

```python
m_model.fit(X_train,y_train)
```

```python
g_model.score(X_test,y_test)
```

```python
m_model.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(g_model, X_train,y_train, cv=5)
```

```python
cross_val_score(m_model,X_train,y_train, cv=5)
```
