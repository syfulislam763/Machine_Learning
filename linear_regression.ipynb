```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("./resource/homeprices.csv")
df
```

```python
%matplotlib inline
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(df.area, df.price, color='red')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
d = pd.read_csv("./resource/areas.csv")
d
```

```python
pr = reg.predict(d)
pr
```

```python
d['price'] = pr
d
```

```python
d.to_csv("./resource/prediction.csv")
```

```python
plt.xlabel('Area', fontsize=20)
plt.ylabel('Price', fontsize='20')
plt.scatter(df.area, df.price)
plt.plot(df.area, reg.predict(df[['area']]), color='green')
```
