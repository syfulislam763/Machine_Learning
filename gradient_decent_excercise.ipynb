```python
import pandas as pd
from sklearn import linear_model
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def gradient_decent(x,y):
    m = 0
    b = 0
    n = len(x)
    learing_rate = 0.0002
    iterations = 1000
    plt.scatter(x, y, color='red', marker='+')
    for i in range(iterations):
        y_predicted = m*x+b
        plt.plot(x, y_predicted, color='green')
        md = -(2/n)*sum(x*(y-y_predicted))
        bd = -(2/n)*sum(y-y_predicted)
        m = m-learing_rate*md
        b = b-learing_rate*bd

    print("m {} b {} learning_rate {}".format(m, b, learing_rate))
```

```python
df = pd.read_csv("./resource/test_scores.csv")
df
```

```python
x = np.array(df['math'])
y = np.array(df['cs'])
```

```python
gradient_decent(x,y)
```
